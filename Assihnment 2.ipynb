```python
import re
from collections import Counter
from nltk.stem import PorterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
```

Create Text

```python
text = """My name is Uswa Aslam.
My father name is Aslam.
I have bachelors in Mathematics and currently i am a student of Data Science.
I have interest in Machine Learning and Deep Learning and Natural Language Processing.
I like coding and projects of ML\\DL and NLP."""
```

Text Writing

```python
with open("test.txt", "w") as file:
    file.write(text)
```

Text Reading

```python
with open("test.txt", "r") as file:
    data = file.read()

print(data)
```

Count of Unique Words

```python
words = data.split()
unique_words = set(words)
```

```python
print("Count of unique words:", len(unique_words))
```

Unique Words & Their Frequencies

```python
word_frequency = Counter(words)

print("Unique words and frequencies:")
print(word_frequency)
```

Convert to Lower Case & Count Unique Words

```python
data_lower = data.lower()
words_lower = data_lower.split()

unique_lower = set(words_lower)
print("Unique words after lower case:", len(unique_lower))
```

Lower Case Words & Their Frequencies

```python
lower_frequency = Counter(words_lower)

print("Lower case word frequencies:")
print(lower_frequency)
```

Sort by Frequency & Save to freq.txt

```python
sorted_freq = lower_frequency.most_common()

with open("freq.txt", "w") as file:
    for word, freq in sorted_freq:
        file.write(word + " : " + str(freq) + "\n")
```

Sort Alphabetically & Save to words.txt

```python
sorted_alpha = sorted(lower_frequency.items())

with open("words.txt", "w") as file:
    for word, freq in sorted_alpha:
        file.write(word + " : " + str(freq) + "\n")
```

Remove Numbers Using Regular Expression

```python
clean_text = re.sub(r'\d+', '', data_lower)

print("Text after removing numbers:")
print(clean_text)
```

Lemmatization

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = []

for word in clean_text.split():
    lemmatized_words.append(lemmatizer.lemmatize(word))

print("Lemmatized words:")
print(lemmatized_words)
```

Stemming

```python
stemmer = PorterStemmer()
stemmed_words = []

for word in clean_text.split():
    stemmed_words.append(stemmer.stem(word))

print("Stemmed words:")
print(stemmed_words)
```

Count Vectorization Construction

```python
count_vectorizer = CountVectorizer()
count_matrix = count_vectorizer.fit_transform([clean_text])

print("Count Vector Matrix:")
print(count_matrix.toarray())
```

```python
print("Feature Names:")
print(count_vectorizer.get_feature_names_out())
```

application of TF-IDF Score

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform([clean_text])

print("TF-IDF Matrix:")
print(tfidf_matrix.toarray())
```

```python
print("Feature Names:")
print(tfidf_vectorizer.get_feature_names_out())
```

Appliation of Naïve Bayes Classifier

```python
documents = [
    "I love machine learning",
    "Deep learning and AI are powerful",
    "I dislike boring subjects",
    "This project is difficult"
]
labels = [1, 1, 0, 0] 
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

model = MultinomialNB()
model.fit(X, labels)

test_sentence = ["I like natural language processing"]
test_vector = vectorizer.transform(test_sentence)

prediction = model.predict(test_vector)
print("Prediction:", prediction)
```
